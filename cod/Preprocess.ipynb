```python

# URLs of the two files
url_123 = "https://github.com/ACS-SII/proiect-2024-sentiment-analysis-study/raw/main/new_set/reviews_negative_123.xlsx"
url_45 = "https://github.com/ACS-SII/proiect-2024-sentiment-analysis-study/raw/main/new_set/reviews_positive_45.xlsx"

# Read the files into pandas dataframes
df_123 = pd.read_excel(url_123)
df_45 = pd.read_excel(url_45)

# Function to split the dataframe into separate subsets by rating (1 to 5)
def split_by_rating(df):
    return {rating: df[df['rating'] == rating] for rating in [1, 2, 3, 4, 5]}

# Split both datasets by rating
ratings_split_123 = split_by_rating(df_123)
ratings_split_45 = split_by_rating(df_45)

# Merge reviews for each rating across both datasets
merged_reviews = {}

for rating in [1, 2, 3, 4, 5]:
    # Get the reviews for each rating from both datasets
    reviews_123 = ratings_split_123.get(rating, pd.DataFrame())['body'].tolist()
    reviews_45 = ratings_split_45.get(rating, pd.DataFrame())['body'].tolist()

    # Merge them into a single list
    merged_reviews[rating] = reviews_123 + reviews_45

# Write the reviews for each rating into separate JSON files
for rating, reviews in merged_reviews.items():
    with open(f"reviews_rating_{rating}.json", "w", encoding="utf-8") as file:
        json.dump(reviews, file, ensure_ascii=False, indent=4)  # Save as a JSON array

print("JSON files for each rating have been generated!")
```

```python
def filter_alphabetic_tokens(tokens):
    # Define the Romanian alphabet range
    romanian_alphabet = "aăâbcddeefghiîjklmnopqrsștțuvwxyâăîşţ"

    # Build the regex pattern for filtering
    pattern = re.compile(f"[^{romanian_alphabet}]")

    # Filter out tokens that contain non-alphabetic characters
    filtered_tokens = [token for token in tokens if not pattern.search(token)]

    return filtered_tokens
```

```python
#voc create
def clean_line_prep(line,voc_nlem,voc):
  rev = line.lower()
  rev = re.sub(r'[.,!?()"\'/:;/\\\-]', ' ', rev)
  rev = re.sub(r'\s+', ' ', rev)
  words=rev.split()
  words = filter_alphabetic_tokens(words)
  voc_nlem.update(words)
  rev= ' '.join(words)
  doc = nlp(rev)
  lemmatized_tokens = [word.lemma for sent in doc.sentences for word in sent.words]
  lemmatized_tokens=[token for token in lemmatized_tokens if isinstance(token, str)]
  lemmatized_tokens=[token for token in lemmatized_tokens if not token.isnumeric()]
  lemmatized_tokens=[token for token in lemmatized_tokens if not any(char.isdigit() for char in token)]
  voc.update(lemmatized_tokens)
  rev=' '.join(lemmatized_tokens)
  return rev
```

```python
def clean_set_prep(dset):
  voc=Counter()
  voc_nlem=Counter()
  cleaned=list()
  for line in dset:
    rev=clean_line_prep(line,voc_nlem,voc)
    cleaned.append(rev)
  return cleaned, voc, voc_nlem
```

```python

# Process and save for each rating
for rating, reviews in merged_reviews.items():
    reviews = [rev for rev in reviews if isinstance(rev, str)]  # Keep only string entries

    # Apply clean_set_prep to each set of reviews
    cleaned, voc, voc_nlem = clean_set_prep(reviews)

    # Save cleaned reviews to a JSON file
    with open(f"cleaned_reviews_rating_{rating}.json", "w", encoding="utf-8") as file:
        json.dump(cleaned, file, ensure_ascii=False, indent=4)

    # Save vocabulary (voc) to a JSON file
    with open(f"voc_rating_{rating}.json", "w", encoding="utf-8") as file:
        json.dump(voc, file, ensure_ascii=False, indent=4)

    # Save non-lemmatized vocabulary (voc_nlem) to a JSON file
    with open(f"voc_nlem_rating_{rating}.json", "w", encoding="utf-8") as file:
        json.dump(voc_nlem, file, ensure_ascii=False, indent=4)

print("Cleaned sets and vocabularies have been saved for each rating!")
```
